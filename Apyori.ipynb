```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('/content/GroceryStoreDataSet (1).csv', names = ['products'], header = None)
```

```python
df.head()
```

```python
df.shape
```

 //..# เนื่องจากข้อมูลของเรารวมกัน เราจะแยกตาม "," ..//

```python
data = list(df["products"].apply(lambda x:x.split(",") ))
data
```

```python
type(data)
```

```python
df.sum()
```

ในการวิเคราะห์การเชื่อมโยง dataframe ต้องเป็น 1 และ 0 หรือประเภทโครงสร้างข้อมูลบูลีนเป็น True / False

```python
from mlxtend.preprocessing import TransactionEncoder
```

// # กระบวนการเข้ารหัสแบบ One Hot เสร็จสิ้นแล้ว /





```python
temp = TransactionEncoder()
temp_df = temp.fit(data).transform(data)
new_df = pd.DataFrame(temp_df,columns=temp.columns_)
new_df
```

// ตอนนี้ Dataframe เหมาะสำหรับการวิเคราะห์ความสัมพันธ์ ขั้นตอนต่อไปคือการคำนวณและตีความค่าการสนับสนุนและค่าความเชื่อมั่น //

// # เราตั้งค่าการสนับสนุนของเราเป็น 20% //

```python
support_values = apriori(new_df, min_support=0.20, use_colnames=True, )
support_values
```

// # เราจัดอันดับข้อมูลของเราตามค่าที่ลดลงสำหรับผู้สนับสนุน //

```python
support_values.sort_values(by = "support", ascending = False)
```

// เราเข้าถึงค่า "ความเชื่อมั่นและการเพิ่ม" โดยใช้ฟังก์ชันassociation_rules //

```python
rules = association_rules(support_values, metric= "confidence", min_threshold = 0.5)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
rules
```

//
มูลค่าการสนับสนุน -> แสดงให้เห็นว่าอัตราการเห็น MAGGI และ TEA ร่วมกันในการซื้อทั้งหมดคือ 0.2

ค่าความเชื่อมั่น -> แสดงให้เห็นว่า 80% ของลูกค้าที่ซื้อ MAGGI ซื้อ TEA ด้วย

มูลค่าที่เพิ่มขึ้น -> แสดงว่ายอดขายของ (TEA) เพิ่มขึ้น 2.28 สำหรับการซื้อกับ MAGGI
